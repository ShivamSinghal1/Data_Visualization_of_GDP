```python
## Import Python Libraries that is going to be used in this project 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Read all data of CSV file
gdp_growth = pd.read_csv('./DataSet/gdp.csv')
gdp_price = pd.read_csv('./DataSet/gdp_cur.csv')
gdp_per_capita = pd.read_csv('./DataSet/gdp_cap.csv')
```

```python
## List of countries for which data is to be analyzed
X = np.array(list(gdp_growth.columns)[1:-5])
country = ["United States","China, People's Republic of","Japan","Germany","India","United Kingdom"]
## Marker to denote different country
marker = ['.', 'o' , '^', '*' , 's' , '.']
```

```python
## To get index according to the country name
index = []
ind = []
for c in country: 
    index.append(gdp_growth.loc[gdp_growth["Real GDP growth (Annual percent change)"] == c].index.values)
for i in index:
    ind.append(i[0])

## To get array of the gdp values of all required country
arr_gdp_growth = gdp_growth.iloc[ind,1:-5].values
arr_gdp_price = gdp_price.iloc[ind,1:-5].values
arr_gdp_per_capita = gdp_per_capita.iloc[ind,1:-5].values

## Y axis of all the country 
Y_gdp_growth = []
for gdp in arr_gdp_growth:
    Y_gdp_growth.append(np.array(gdp,dtype='float32'))
Y_gdp_price = []
for gdp in arr_gdp_price:
    Y_gdp_price.append(np.array(gdp,dtype='float32'))
Y_gdp_per_capita = []
for gdp in arr_gdp_per_capita:
    Y_gdp_per_capita.append(np.array(gdp,dtype='float32'))
    
plt.style.use('seaborn')
plt.figure(figsize = (20,10))
for i in range(len(Y_gdp_growth)):
    plt.plot(X,Y_gdp_growth[i],label=country[i],marker= marker[i])
plt.legend(loc = 'upper left',fontsize = 16)
plt.xlabel("Years",fontsize = 20)
plt.ylabel("Real GDP growth (Annual percent change)",fontsize = 20)
plt.title("Real GDP growth (Annual percent change) vs Years for Top 5 countries in GDP",fontsize = 20)
plt.savefig('gdp_growth.png')
plt.show()
plt.figure(figsize = (20,10))
for i in range(len(Y_gdp_price)):
    plt.plot(X,Y_gdp_price[i],label=country[i],marker= marker[i])
plt.legend(loc = 'upper left',fontsize = 16)
plt.xlabel("Years",fontsize = 20)
plt.ylabel("GDP, current prices (Billions of U.S. dollars)",fontsize = 20)
plt.title("GDP, current prices (Billions of U.S. dollars) vs Years for Top 5 countries in GDP",fontsize = 20)
plt.savefig('gdp_current_prices.png')
plt.show()
plt.figure(figsize = (20,10))
for i in range(len(Y_gdp_per_capita)):
    plt.plot(X,Y_gdp_per_capita[i],label=country[i],marker= marker[i])
plt.legend(loc = 'upper left',fontsize = 16)
plt.xlabel("Years",fontsize = 20)
plt.ylabel("GDP per capita, current prices (U.S. dollars per capita)",fontsize = 20)
plt.title("GDP per capita, current prices (U.S. dollars per capita) vs Years for Top 5 countries in GDP",fontsize = 20)
plt.savefig('gdp_per_capita.png')
plt.show()
```
